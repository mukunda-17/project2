```python
!pip install spacy nltk textblob pandas -q
!python -m spacy download en_core_web_sm -q

import pandas as pd
import re
import nltk
import spacy

from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

nlp = spacy.load("en_core_web_sm")

data = {
    "ticket": [
        "Hii!! My payment failed twice while ordering on ur app!!!",
        "I cant login to my accnt since yesterday :(",
        "Refund not recevied after cancellation.",
        "Delivery delayed by 5 days and customer support not responding.",
        "Need help resetting my password ASAP!!!"
    ]
}

df = pd.DataFrame(data)

def clean_text(text):

    text = text.lower()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(r"[^a-zA-Z\s]", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    return text

def correct_spelling(text):

    corrected = TextBlob(text).correct()

    return str(corrected)

stop_words = set(stopwords.words('english'))

def preprocess_text(text):

    tokens = word_tokenize(text)

    filtered_tokens = [
        word for word in tokens
        if word not in stop_words
    ]

    doc = nlp(" ".join(filtered_tokens))

    lemmatized = [
        token.lemma_
        for token in doc
    ]

    return " ".join(lemmatized)

def extract_entities(text):

    doc = nlp(text)

    entities = []

    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    return entities

df["cleaned_text"] = df["ticket"].apply(clean_text)

df["corrected_text"] = df["cleaned_text"].apply(correct_spelling)

df["processed_text"] = df["corrected_text"].apply(preprocess_text)

df["entities"] = df["ticket"].apply(extract_entities)

labels = [
    "Payment Issue",
    "Login Issue",
    "Refund Issue",
    "Delivery Issue",
    "Password Reset"
]

df["label"] = labels

df.to_csv("processed_support_tickets.csv", index=False)

print("Dataset Saved Successfully\n")

print(df)
```
